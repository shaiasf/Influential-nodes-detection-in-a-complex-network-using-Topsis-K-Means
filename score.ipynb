```python
g = nx.read_edgelist('fbData.txt',create_using=nx.Graph(),nodetype=int)
```

```python
#centrality measuers computing 
dc = nx.degree_centrality(g)    
bc = nx.betweenness_centrality(g)
cc = nx.closeness_centrality(g)
ec = nx.eigenvector_centrality(g)
```

```python
evolution_matrix = fcts.creat_evaluation_matrix(g,dc,bc,cc,ec)#evolution matrix is evaluation matrix only as a dataFrame 
ev_matrix_np = fcts.evolution_matrix_tonp(evolution_matrix)#numpy array version of evaluation matrix
weights = fcts.entropy(g=g,ev_mat=ev_matrix_np)
t = fcts.Topsis(ev_matrix_np,weights,criteria = np.array([True,True,True,True]))
t.calc()
top_k_noeuds, tous_les_noeuds = fcts.top_meth(evolution_matrix,t=t)
```

```python
import pandas as pd
kinit = pd.read_csv('kinit.csv')
kinit = kinit.drop(columns = ['Unnamed: 0'], axis=1)

kinit
```

```python
#score = 0.3dc + 0.2bc + 0.2cc + 0.3ec
score = []
for idx, row in kinit.iterrows():
  score_value = 0.33*row['dc'] + 0.33*row['bc'] + 0.34*row['cc']
  score.append(score_value)

kinit['score'] = score
```

```python
kinit
```

```python
kinit = kinit.sort_values(by='score',ascending=False)
kinit
```

```python
import math 

def notre_kmeans(dataframe, k = 10):


    data = dataframe.values
    n = data.shape[0]
    # select the top k points (highest scores) as centroids
    centroids = data[:k]
    labels = np.zeros(n)

    # Assign each data point to the closest cluster center
    #new_labels = np.array([np.argmin([np.abs(data[i]-cent) for cent in centroids]) for i in range(n)])
    while True: 

        new_labels = np.array([np.argmin([np.sqrt(np.sum((data[i]-cent)**2)) for cent in centroids]) for i in range(n)])
        labels = new_labels
        if (new_labels == labels).all():
            break
        labels = new_labels
        # Update cluster centers
        for i in range(k):
            centroids[i] = np.mean(data[labels == i], axis = 0)


    return labels, centroids
    



lbs, cents = notre_kmeans(kinit[['score', 'C']],10)
cents
```
